# Espectros de las muestras y sus formas de onda

```python
# imports
import matplotlib.pyplot as plt
import numpy as np
import wave, sys
# muestra la onda sonora
def visualize(path: str):

  # abriendo el archivo .wav
  raw = wave.open(path)
  
  # lee todos los fotogramas
  # -1 indica todos o maximos fotogramas
  signal = raw.readframes(-1)
  signal = np.frombuffer(signal, dtype ="int16")
  
  # obtiene la frecuencia de los fotogramas
  f_rate = raw.getframerate()

  # para graficar el eje x en segundos
  # se necesita la frecuencia
  # y dividirla por el tamaño de la señal
  # para crear un vector de tiempo
  # linealmente espaciado con el tamaño
  # del archivo de audio
  time = np.linspace(
    0, # inicio
    len(signal) / f_rate,
    num = len(signal)
  )
  # usando matplotlib para graficar
  # crea nueva figura
  plt.figure(1)
  
  # titulo de la gráfica
  plt.title("Forma de Onda de muestra1.wav")
  
  # leyenda del eje x
  plt.xlabel("Tiempo")

  # leyenda del eje y
  plt.ylabel("Amplitud")
  
  # graficando
  plt.plot(time, signal)
  
  # muestra el gráfico
  # en nueva ventana
  plt.show()


if __name__ == "__main__":

  # obtiene el valor de la linea de comandos
  path = '/content/muestra1.wav'

  visualize(path)
```

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# Se lee el archivo .wav
sample_rate, data = wavfile.read('muestra1.wav')

# Se extrae el canal de audio si es estéreo
if data.ndim > 1:
    data = data[:, 0]  # Toma el primer canal

# FFT (Transformada Rápida de Fourier)
fft_result = np.fft.fft(data)

# Se obtienen las frecuencias correspondientes para graficar
frequencies = np.fft.fftfreq(len(fft_result), d=1/sample_rate)

# Se grafica el espectro
plt.figure(figsize=(10, 4))
plt.plot(frequencies[:len(frequencies)//2], np.abs(fft_result[:len(frequencies)//2].real))
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Amplitud')
plt.title('Espectro de muestra1.wav')
plt.xlim(0,sample_rate/2)
plt.ylim(0,)
plt.grid(True)
plt.show()
```

# Filtrando las señales

```python
from scipy.io import wavfile
from scipy import signal

# Se lee el archivo .wav
sample_rate, data = wavfile.read('muestra1.wav')

# Se extrae el canal de audio si es estéreo
if data.ndim > 1:
    data = data[:, 0]  # Toma el primer canal

# Define la frecuencia de corte para el filtro lowpass (en Hz)
lp_cutoff_frequency = 4000

# Se diseña el filtro
lp_order = 8  # Orden del filtro (qué tan pronunciada es la pendiente)
lp_b, lp_a = signal.butter(lp_order, lp_cutoff_frequency/(sample_rate/2), btype='lowpass')

# Se aplica el filtro al audio
filtered_data = signal.lfilter(lp_b, lp_a, data)

# Define la frecuencia de corte para el filtro highpass (en Hz)
hp_cutoff_frequency = 600 

# Se diseña el filtro
hp_order = 8  # Adjust this value as per your requirement
hp_b, hp_a = signal.butter(hp_order, hp_cutoff_frequency/(sample_rate/2), btype='highpass')

# Se aplica el segundo filtro
filtered_data = signal.lfilter(hp_b, hp_a, filtered_data)

# Se guarda el archivo filtrado
wavfile.write('filtered1.wav', sample_rate, filtered_data.astype(np.int16))
```
